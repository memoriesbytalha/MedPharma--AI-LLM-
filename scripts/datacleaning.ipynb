```python
csv_file_path = os.path.join(MAIN_DIR, 'data','merged_interactions_new.csv')
df = pd.read_csv(csv_file_path)
```

```python
import pandas as pd

# Define the columns that must not have missing values
important_cols = ['Drug_A', 'Drug_B', 'DrugA_SMILES', 'DrugB_SMILES', 'DrugA_Formula', 'DrugB_Formula']

# Drop rows where any of these columns have NaN
df = df.dropna(subset=important_cols)

# Reset index (optional)
df.reset_index(drop=True, inplace=True)

print("Cleaned DataFrame shape:", df.shape)
df.isna().sum()
```
